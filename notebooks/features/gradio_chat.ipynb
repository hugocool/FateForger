```python
import gradio as gr
from autogen_core import AgentId
from autogen_agentchat.messages import TextMessage
from fateforger.core.runtime import initialize_runtime

async def respond(message, history):
    runtime = await initialize_runtime()
    result = await runtime.send_message(
        TextMessage(content=message, source="gradio"),
        recipient=AgentId("planner_agent", key="default"),
    )
    reply = getattr(result, "chat_message", result)
    return getattr(reply, "content", str(reply))

gr.ChatInterface(fn=respond, title="Planner Agent Chat").launch(inline=True)
```
